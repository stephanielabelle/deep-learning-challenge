## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df["NAME"].value_counts()[application_df['NAME'].value_counts()>200]
```

```python
# Drop the non-beneficial ID columns, 'EIN'.
application_df = application_df.drop(['EIN'], axis=1)
application_df.head()
```

```python
#Using the NAME column to find number of applications by applicant and putting in "NAME_COUNT" column
application_df['NAME_COUNT'] = application_df.groupby(['NAME'])["ASK_AMT"].transform('count')
application_df

```

```python
# Binning the Applicant NAME by Count and putting into column "NAME_COUNT_BIN"
application_df["NAME_COUNT_BIN"] = pd.cut(application_df['NAME_COUNT'], bins=[0,5,10,25,50,75,100,5000], labels=["0-5", "6-10", "11-25", "26-50", "51-75", "76-100", "101+"])
application_df
```

```python
# Finding distribution of applicants in application counts
application_df["NAME_COUNT_BIN"].value_counts()
```

```python
# Removing "NAME_COUNT" and "NAME" columns
application_df = application_df.drop(['NAME', 'NAME_COUNT'], axis=1)
```

```python
# Determine the number of unique values in each column.
print(application_df.nunique())
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df["APPLICATION_TYPE"].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace = application_df["APPLICATION_TYPE"].value_counts()[application_df['APPLICATION_TYPE'].value_counts()<100].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df["CLASSIFICATION"].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df["CLASSIFICATION"].value_counts()[application_df['CLASSIFICATION'].value_counts()>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = application_df["CLASSIFICATION"].value_counts()[application_df['CLASSIFICATION'].value_counts()<1000].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
application_df.head()
```

```python
application_df['SPECIAL_CONSIDERATIONS'].value_counts()
```

```python
# Removing SPECIAL_CONSIDERATIONS columns - few entries with 'Y' option
application_df = application_df.drop(['SPECIAL_CONSIDERATIONS'], axis=1)
```

```python
application_df.dtypes
```

```python
application_cat = application_df.dtypes[application_df.dtypes =="object"].index.tolist()
application_cat.append("NAME_COUNT_BIN")
application_df[application_cat].nunique()
```

```python
len(application_cat)
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
col_dummies = pd.get_dummies(application_df[application_cat], dtype=int)
application_df_transformed = pd.concat([application_df, col_dummies], axis=1)
application_df_processed = application_df_transformed.drop(columns=application_cat)
application_df_processed
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df_processed['IS_SUCCESSFUL'].values
X = application_df_processed.drop(['IS_SUCCESSFUL'],axis=1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
number_input_features = len(X_train[0])
number_input_features
```

```python
len(X_train)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 50
hidden_nodes_layer2 = 30
#Adding third layer for optimization
hidden_nodes_layer3 = 10



nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save("AlphabetSoupCharity_Optimization.h5")
```
